```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def activation_function(x):
    result=np.zeros(len(x))
    for i in range(len(x)):
        
        if x[i] >=0 and x[i]<2 :
            result[i]=x[i];
        if x[i]>=2:
            result[i]=2
        if x[i]<0:
            result[i]=0
    return result
```

```python
def weights(R1,R2,C1,C2):
    result=np.zeros(R2+1)
    result[0:R1+1]=C1
    result[R1+1:R2+1]=C2
    return result
```

```python
def product(w,x):
    res=np.zeros(len(x))
    for i in range(len(w)):
        res[i]=np.dot(w[i,:],x)
    return res
```

```python
def max_net(x,epsilon):
    while(1):
        cnt=0
        w=weights(epsilon,x)
        xnew=product(w,x)
        res=activation_function(xnew)
        for i in range(len(res)):
            if res[i]>0:
                cnt=cnt+1
                index=i+1
        if cnt==1:
            break
        print("updating vector:",res)
        x=res
    return res,index
```

```python
x=[0.27,0.35,0.44,0.58,0.66,0.77,0.4,0.32,0.20,0.15,0.08]
R1=0
R2=5000
C1=0.7
C2=-0.1
```

```python
def Mexican_Hat(R1,R2,C1,C2,x_old,t_max):
    cnt=0
    vector=np.zeros((t_max,len(x_old)))
    while(cnt!=t_max):
        
        xnew=np.zeros(len(x_old))
        for i in range(len(x_old)):
            for k in range(2*R1+1):
                if i+k-R1<=len(x_old)-1 and 0<=i+k-R1 :
                    xnew[i]=xnew[i]+C1*(x_old[i+k-R1])
            
            for k in range(R2-R1):
                if i+k-R2<=len(x_old)-1 and 0<=i+k-R2:
                    xnew[i]=xnew[i]+C2*(x_old[i+k-R2])
            
            for k in range(R2-R1):
                if i+k+R1+1<=len(x_old)-1 and 0<=i+k+R1+1 :
                    xnew[i]=xnew[i]+C2*(x_old[i+k+R1+1])
        res=activation_function(xnew)
        vector[cnt,:]=res
        x_old=res
        cnt=cnt+1
    return res,vector
```

```python
x_old=x
t_max=10
res,vector=Mexican_Hat(R1,R2,C1,C2,x_old,t_max)
```

```python
plt.plot(vector[0],label='t=1')
plt.plot(vector[1],label='t=2')
plt.plot(vector[2],label='t=3')
plt.plot(vector[3],label='t=4')
plt.plot(vector[4],label='t=5')
plt.plot(vector[5],label='t=6')
plt.plot(vector[6],label='t=7')
plt.plot(vector[7],label='t=8')
plt.plot(vector[8],label='t=9')
plt.plot(vector[9],label='t=10')
plt.title("Results for Mexican_Hat example")
plt.ylabel("x value")
plt.xlabel("Index")
plt.legend()
plt.show()
```
